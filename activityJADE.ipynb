> Using a `for` loop, print out the numbers from 1 to 100. If the `number` is divisible by 3, print "`Fizz`". If the `number` is divisible by 5, print "`Buzz`". If the `number` is divisible by both 3 and 5, print "`FizzBuzz`". Otherwise, print the `number`.

```python
for num in range(1,101):

  if (num % 3 ==0):
    print('Fizz')
  if (num % 5 == 0):
    print('Buzz')
  if (num % 3 == 0) and (num % 5 == 0):
    print('FizzBuzz')    
  else:
    print(num)     
```
